```
import pandas as pd

cumhuriyet_meta = pd.read_csv('suder/suder-v1/meta_cumhuriyet.csv')
cumhuriyet_meta.head(2)
```

```
sabah_meta = pd.read_csv('suder/suder-v1/meta_sabah.csv')
sabah_meta.head(2)
```

```
import json

with open('suder/suder-v1/texts_sabah.json') as json_file:
    texts_sabah = json.load(json_file)

with open('suder/suder-v1/texts_cumhuriyet.json') as json_file:
    texts_cumhuriyet = json.load(json_file)

    
```

```
texts_sabah['0']
```

```
sabah = sabah_meta[['TextId','Category']]
sabah['text'] = sabah.TextId.apply(lambda x:texts_sabah[str(x)] )
sabah.head(5)
```

```
cumhuriyet = cumhuriyet_meta[['TextId','Category']]
cumhuriyet['text'] = cumhuriyet.TextId.apply(lambda x:texts_cumhuriyet[str(x)] )
cumhuriyet.head(5)
```

```
# get class_dict from dataframe
# cumhuriyet_class_dict = {category:i for i,category in enumerate(cumhuriyet.Category.unique())}
# sabah_class_dict = {category:i for i,category in enumerate(sabah.Category.unique())}
```

```
sabah_class_dict = {'ekonomi': 2, 'gundem': 1, 'yasam': 0, 'yazarlar': 3}
cumhuriyet_class_dict = {'cevre': 12, 'dunya': 8, 'egitim': 6, 'ekonomi': 10, 'foto': 1, 'kultursanat': 7, 'saglik': 11, 'siyaset': 3, 'spor': 5, 'teknoloji': 13, 'turkiye': 4, 'video': 2, 'yasam': 9, 'yazarlar': 0}
```

```
sabah['label'] = sabah.Category.apply(lambda x:int(sabah_class_dict[x]) )
sabah['text'] = sabah['text'].apply(lambda x: x.replace('\n', ' '))
sabah = sabah[['text','label']]
sabah
```

```
cumhuriyet['label'] = cumhuriyet.Category.apply(lambda x:int(cumhuriyet_class_dict[x]) )
cumhuriyet['text'] = cumhuriyet['text'].apply(lambda x: x.replace('\n', ' ').strip())
cumhuriyet = cumhuriyet[['text','label']]

# drop rows with no text
cumhuriyet = cumhuriyet.loc[cumhuriyet['text'] != '']
cumhuriyet
```

```
[str(x) for x in cumhuriyet.label.value_counts().keys()]
```

```
# finding how many words is there in the longest news article
sabah_max_len = 0
for news_text in sabah.text.values:
  n_tokens = len(news_text.split())
  if(sabah_max_len < n_tokens ):
    sabah_max_len = n_tokens
sabah_max_len
```

```
# finding how many words is there in the longest news article
cumhuriyet_max_len = 0
for news_text in cumhuriyet.text.values:
  n_tokens = len(news_text.split())
  if(cumhuriyet_max_len < n_tokens ):
    cumhuriyet_max_len = n_tokens

cumhuriyet_max_len
```

```
sabah.drop_duplicates(subset='text', keep="last",inplace=True)
cumhuriyet.drop_duplicates(subset='text', keep="last",inplace=True)
```

```
from sklearn.model_selection import train_test_split

train_sabah, test_sabah = train_test_split(sabah, test_size=0.2,random_state=3)
```

```
from sklearn.model_selection import train_test_split

train_cumh, test_cumh = train_test_split(cumhuriyet, test_size=0.2,random_state=3)
```
